## Jointure des dataframes

## Nettoyage des données

Afin d’obtenir un jeu de données exploitable, je vérifie les anomalies qui peuvent fausser les analyses. Je trouve deux anomalies comme suivant:
 
 - Anomalie concernant les clients de test
 
 Il s'agit des tests, ce ne sont pas les produits que la librairie a vendus, inutile pour l'analyse.
 
 
 - Anomalie concernant le produit de la référence 0_2245
 
 Il s'agit d'un produit de la catégorie 0 que la librairie a vendu, mais il manque du prix unitaire, après vérifier la variance, qui est égale 24, je trouve que c'est grand, c'est mieux de ne pas imposer par le prix moyen de cette catégorie.
 
Donc, je décide de supprimer les données de ces deux anomalies.
 

### Anomalie - clients de test

```python
# Anomalie détectée pour id_prod = 'T_0', il s'agit des tests

cherche1 = df1[df1['id_prod'] == 'T_0']
cherche1
```

```python
# Vérification s'il y a de fause transaction réalisée et mal enregistrée par ces deux clients de test
# Il y en a pas, car le nombre de ligne est toujours de 200

test = df1[(df1['client_id'] == 'ct_0') | (df1['client_id'] == 'ct_1')]
test
```

```python
# Transformation de la colonne date au format datetime

df1['date'] = pd.to_datetime(df1['date'], errors='coerce')
```

```python
df1.info()
```

```python
# Revérification pour la référence 'T_0'
# La colonne date pour cette réf est bien transformet en NaT

cherche2 = df1[df1['id_prod'] == 'T_0']
cherche2
```

### Anomalie - produit référence 0_2245

```python
# Recherche des valeurs manquantes qui correspond à la différence que l'on a trouvé précedent, ref '0_2245'
# id_prod commence par 0, il s'agit de la catégorie 0

cherche3 = df1[df1['id_prod'] == '0_2245']
cherche3
```

```python
# Calcul de la variance
# Si la variance n'est pas énorme, la solution pour nettoyer cette anomalie sera de l'imputation par la moyenne
# Dans notre cas, 24.39 pour la catégorie 0 pour un livre, je trouve que c'est beaucoup

df1.groupby('categ').var()['price']
```

### Solution - suppression

```python
# Nettoyage des données, solution choisie : supprimer

data_propre = df1.dropna(how="any").reset_index(drop=True)
data_propre
```

```python
# Répartition du CA par client pour les ventes B to C

CA_par_client = vente_toC.groupby('client_id').agg({'price': 'sum'}).sort_values(
    'price', ascending=False).reset_index()
CA_par_client
```

```python
# Courbe de Lorenz sans les 4 gros clients
# La répartition du CA par client n'est pas égalité, il y a des clients achètent plus, autres achètent moins

variable = CA_par_client['price'].values
lorenz = np.cumsum(np.sort(variable)) / variable.sum()
lorenz = np.append([0],lorenz)
fig, ax = plt.subplots(figsize=[10, 10])
ax.plot(np.arange(lorenz.size)/(lorenz.size-1), lorenz, label = 'Courbe de Lorenz')
ax.plot([0, 1], [0, 1], color='k', label = 'Première bissectrice')
plt.xlabel('Part cumulée en % des clients')
plt.ylabel('Part cumulée en % du CA')
plt.title('Répartition du CA par client hors les 4 clients')
plt.legend(loc='upper left')
plt.show()
```

```python
# Calcul de l'indice de Gini

AUC = (lorenz.sum() -lorenz[-1]/2 -lorenz[0]/2)/lorenz.size
S = 0.5 - AUC
gini = 2*S
gini
```

## Lien entre le genre d’un client et les catégories des livres achetés 

En regardant les valeurs du tableau de contingence, nous pouvons constater qu'il y a plus de ventes chez les femmes.
Nous pouvons, au premier abord, constater qu'il y a une différence de comportement entre les genres.

Cependant, lorsque nous regardons en termes de proportion de chaque catégorie, cette différence de comportement entre les genres semblent minimes voir négligeable et peut potentiellement être attribué à une fluctuation statistique.

La conclusion que je peux tirer selon le test de Khi2 et le heatmap est donc que:

Le choix d'un produit ne dépend pas du sexe des clients, mais dépend de la catégorie des produits. Le genre d’un individu ne permet pas de prédire si l'achat se portera plus sur une catégorie qu'une autre. Un individu aura donc une certaine chance de choisir une catégorie plus qu'une autre indépendamment du genre.

### Création d'une table de contingence

```python
# Création de la table contingence - variables « sex » et « catégorie »

x = 'sex'
y = 'categ'
table = vente_toC[[x, y]].pivot_table(index=x, columns=y,
                                      aggfunc=len, margins=True, margins_name='Total')
table
```

```python
pd.crosstab(vente_toC['sex'], vente_toC['categ'], normalize='index')
```

### Test de Khi2

```python
# Test de Khi2 - variables  « genre des clients » et « catégories des livres achetés »

# H0: le choix d'un produit ne dépend pas de ni la catégorie, ni la sex des clients
# H1: le choix d'un produit dépend soit de la sex des clients, soit de la catégorie
# Test au seuil de 5%
# Rejetons l’hypothèse nulle au seuil de 5%, donc pour une pvaleur 0.002 inférieure à 0.05
# Le choix d'un produit dépend soit de la sex des clients, soit de la catégorie

chi2, pvalue, degrees, expected = stats.chi2_contingency(table)
chi2, degrees, pvalue
```

```python
# Pour vérifier la test de Khi2, on va regarder à l'aide de heatmap
# En effet, le choix d'un produit ne dépend pas de la sex des clients, mais dépend de la catégorie des produits

tx = table.loc[:, ["Total"]]
ty = table.loc[["Total"], :]
n = len(vente_toC)
indep = tx.dot(ty) / n

c = table
measure = (c-indep)**2/indep
xi_n = measure.sum().sum()
table1 = measure/xi_n

sns.set(rc={'figure.figsize': (16, 10)},font_scale=1.4)
sns.heatmap(table1.iloc[:-1, :-1], annot=c.iloc[:-1, :-1], fmt="d")
plt.show()
```

```python
# Montant total des achats par âge des clients

plt.figure(figsize=(15, 10))
plt.scatter(table2['age'],table2['sum'])
plt.ylabel('Montant total des achats')
plt.xlabel('Âge des clients')
plt.title("Montant total des achats par âge des clients")
plt.show()
```

## Lien entre l’âge des clients et les catégories des livres achetés

### Test de Kolmogorov-Smirnov

```python
# Vérification la normalité de chaque group de la variable 'categ'

# H0:la variable suivre la loi normale
# H1:la variable ne suivre pas la loi normale

# P=0.0, inférieur au seuil de 0.05
# Rejetons l’hypothèse nulle au seuil de 5%
# La variable testée ne suivre pas la loi normale
# On doit utiliser les tests non paramétriques

x1 = vente_toC[vente_toC['categ'] == 0.00]['age']
resultat = stats.kstest(x1, 'norm')
resultat
```

```python
x2 = vente_toC[vente_toC['categ'] == 1.00]['age']
resultat = stats.kstest(x2, 'norm')
resultat
```

```python
x3 = vente_toC[vente_toC['categ'] == 2.00]['age']
resultat = stats.kstest(x3, 'norm')
resultat
```

### Test de Levene

```python
# Vérification l'homogénéité de chaque groupe de la variable 'categ'

# H0:les variances de chaque groupe de la variable 'categ' sont égaux
# H1:les variances de chaque groupe de la variable 'categ' ne sont pas égaux

# P=0.0, inférieur au seuil de 0.05
# Rejetons l’hypothèse nulle au seuil de 5%
# Les variances de chaque groupe de la variable 'categ' ne sont pas égaux
# On doit utiliser les tests non parametriques

x = vente_toC[vente_toC['categ'] == 0.00]['age']
y = vente_toC[vente_toC['categ'] == 1.00]['age']
z = vente_toC[vente_toC['categ'] == 2.00]['age']
stat, p = stats.levene(x, y, z)
print(stat, p)
```

```python
# Test de Kruskal-Wallis - variables  « âge » et « catégories des livres achetés »

# H0: La répartition des âges des clients dans les trois catégories sont égaux
# H1: La répartition des âges des clients dans les trois catégories ne sont pas égaux
# Test au seuil de 5%
# Rejetons l’hypothèse nulle au seuil de 5%, donc pour une pvaleur < 0.05
# La répartition des âges des clients dans les trois catégories ne sont pas égaux

stats.kruskal(x, y, z)
```
