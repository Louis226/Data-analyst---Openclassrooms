# Importation des données

```python
import pandas as pd
```

## Données Population

```python
population = pd.read_csv(
    'Desktop/Openclassrooms/P9_Shibin_YANG/Population_2000_2018.csv')
```

### Nettoyage

```python
# Changement de l'unité de la valeur
population['Population'] = population['Valeur']*1000
population
```

```python
# Sélection des colonnes souhaitées
pop = population[['Code zone', 'Zone', 'Année', 'Population']]
```

```python
# Sélection des données de l'année souhaitée
population2017 = pop[pop['Année'] == 2017].reset_index(drop=True)
```

### Vérification des doublons

## Données Disponibilité alimentaire

```python
dispoAlimentaire = pd.read_csv(
    'Desktop/Openclassrooms/P9_Shibin_YANG/DisponibiliteAlimentaire_2017.csv')
```

```python
dispoAlimentaire
```

```python
dispoAlimentaire.info()
```

### Nettoyage

```python
# Réorganisation des données - disponibilité totale de toutes les alimentaires
dispoTotal = dispoAlimentaire.pivot_table(
    columns='Élément', index=['Code zone'], values='Valeur', aggfunc=sum)
dispoTotal
```

```python
# Sélection des colonnes besoins pour ensuit calculer la proportion de poulet par rapport à l'importation totale et la consommation totale
dispoTotal = dispoTotal[['Importations - Quantité', 'Nourriture']]
dispoTotal
```

```python
# Renommage des colonnes
dispoTotal = dispoTotal.rename(columns={
                               'Importations - Quantité': 'Importations Totale', 'Nourriture': 'Nourriture Totale'})
dispoTotal
```

```python
# Sélection des colonnes souhaitées pour ensuite créer une dataframe qui ne concerne que du poulet
dispoPoulet = dispoAlimentaire[['Code zone',
                                 'Zone', 'Élément', 'Produit', 'Valeur']]
dispoPoulet
```

```python
# Sélection des données en limitant le produit qui ne concernne que du poulet
dispoPoulet = dispoPoulet[dispoPoulet['Produit'] == 'Viande de Volailles'].reset_index(drop=True)
dispoPoulet
```

```python
# Réorganisation des données - disponibilité poulet
dispoPoulet = dispoPoulet.pivot_table(columns='Élément', index=['Code zone'], values='Valeur')
dispoPoulet
```

```python
dispoPoulet.info()
```

```python
# Sélection des colonnes besoins
dispoPoulet = dispoPoulet[['Disponibilité intérieure', 'Exportations - Quantité',
                           'Importations - Quantité', 'Production', 'Nourriture']]
dispoPoulet
```

```python
# Renommage des colonnes
dispoPoulet = dispoPoulet.rename(columns={'Disponibilité intérieure': 'Dispo intérieure poulet', 
                        'Exportations - Quantité': 'Exportations poulet',
                        'Importations - Quantité': 'Importations poulet', 
                        'Production': 'Production poulet','Nourriture': 'Consommation poulet'})
dispoPoulet
```

```python
# Check le différent entre dispoPoulet et dispoTotal
diff_code_zone = list(set(dispoTotal.index.unique()).difference(
    dispoPoulet.index.unique()))
diff_code_zone
```

### Liste des pays sans info poulet

```python
# Deux pays différents sont trouvés, qui n'ont pas d'info par rapport au poulet
Pays_sans_info1 = population2017.loc[population2017['Code zone'].isin(
    pd.Series(diff_code_zone)), ['Code zone', 'Zone']].reset_index(drop=True)
Pays_sans_info1
```

```python
# Jointure dispoPoulet et dispoTotal pour ensuite faire le calcul
dispoPoulet2017 = pd.merge(dispoTotal, dispoPoulet, how='inner', on='Code zone')
dispoPoulet2017
```

```python
# Changement de l'unité des valeurs
dispoPoulet2017 = dispoPoulet2017.multiply(1000)
dispoPoulet2017
```

```python
# Mise à jour de l'index
dispoPoulet2017.reset_index(inplace=True)
dispoPoulet2017
```

```python
dispoPoulet2017.info()
```

## Calcul des variables

```python
# Calcul de la proportion de poulet par rapport à l'importation totale et la consommation totale
dispoPoulet2017['Part importation poulet(%)'] = dispoPoulet2017['Importations poulet'] / \
    dispoPoulet2017['Importations Totale']*100
dispoPoulet2017['Part consommation poulet(%)'] = dispoPoulet2017['Consommation poulet'] / \
    dispoPoulet2017['Nourriture Totale']*100
dispoPoulet2017
```

```python
# Arrondissement à deux décimales
dispoPoulet2017 = dispoPoulet2017.round({'Part importation(%)': 2, 'Part nourriture(%)': 2})
dispoPoulet2017
```

```python
# Suppression des colonnes inutiles
dispoPoulet2017 = dispoPoulet2017.drop(
    columns=['Importations Totale', 'Nourriture Totale'])
dispoPoulet2017
```

### Vérification des doublons

```python
# Vérification des doublons pour la colonne Code zone

check1 = dispoPoulet2017.duplicated('Code zone')
check1
```

```python
# Pas de doublons trouvés

check1[check1 == True]
```

```python
# Recherche de valeurs différentes pour la colonne Code zone entre deux dataframes - population2017 et dispoPoulet2017

diff_code_zone1 = list(set(population2017['Code zone'].unique()).difference(
    dispoPoulet2017['Code zone'].unique()))
diff_code_zone1
```

### Liste des pays sans info poulet

```python
# Liste des pays qui n'ont pas d'info par rapport au poulet
Pays_sans_info2 = population2017.loc[population2017['Code zone'].isin(
    pd.Series(diff_code_zone1)), ['Code zone', 'Zone']].reset_index(drop=True)
Pays_sans_info2
```

### Jointure des dataframes

```python
# Jointure des dataframes
df = pd.merge(population2017, dispoPoulet2017, how='inner', on='Code zone')
df
```

```python
df.info()
```
